```python
os.environ["CUDA_VISIBLE_DEVICES"]="-1"
path_to_artifacts = "./mlruns/0/f3c93f9bcc044c85be04a157f6ea2526/artifacts"
path_to_artifacts_old1000 = "./mlruns/0/91b9a22514cc41d0960060d114e56037/artifacts"
path_to_artifacts_1 = "./mlruns/0/2d15453eb121486a89be229900851652/artifacts"

with open(f'{path_to_artifacts}/input_cfg/metric_names.json') as f:
        metric_names = json.load(f)
path_to_model = f'{path_to_artifacts}/model'
model = load_model(path_to_model, {name: lambda _: None for name in metric_names.keys()}) # workaround to load the model without loading metric functions
```

```python

print("Tau dense kernel weights:\n")
weights = model.get_layer("tau_dense_1").weights
kernel_weights = weights[0].numpy()
#print(kernel_weights)

#average
print("Averages:")
avg = np.zeros(len(kernel_weights[:, 0]))
min = np.zeros(len(kernel_weights[:, 0]))
max = np.zeros(len(kernel_weights[:, 0]))
zeros = np.zeros(len(kernel_weights[:, 0]))
for i in range(len(avg)):
    avg[i] = np.mean(np.abs(kernel_weights[i, :]))
    min[i] = np.min(np.abs(kernel_weights[i,:]))
    max[i] = np.max(np.abs(kernel_weights[i,:]))
    zeros[i] = len(kernel_weights[i,:]) -np.count_nonzero(kernel_weights[i,:])
print(avg)
print("minima:")
print(min)
print("maxima:")
print(max)
print("Non zeros")
print(zeros)

    
    


```
